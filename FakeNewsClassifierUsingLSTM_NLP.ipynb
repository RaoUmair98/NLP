```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/training_data_for_nlp.csv")
df.head()
```

```python
x = df.drop('label', axis=1)
```

```python
y=df['label']
```

```python
from keras.layers import Embedding
from keras.models import Sequential
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import one_hot
from keras.layers import LSTM
from keras.layers import Dense
```

```python
voc_size = 10000
```

```python
mess = x.copy()
```

```python
mess['title'][1]
```

```python
mess.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(mess)):
  print(i)
  review = re.sub('[^a-zA-Z]', ' ', str(mess['title'][i]))
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
hot = [one_hot(words, voc_size) for words in corpus]
hot
```

```python
sent_length=20
embed = pad_sequences(hot, padding='pre',  maxlen= sent_length)
print(embed)
```

```python
embed[0]
```

```python
dim = 40
model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics =['accuracy'])
print(model.summary())
```

```python
len(embed), y.shape
```

```python
import numpy as np
x_final = np.array(embed)
y_final = np.array(y)
```

```python
x_final.shape, y_final.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final, y_final , test_size=0.33, random_state=0)
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=60)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python

accuracy_score(y_test, y_pred)
```
