```python
from keras.preprocessing.text import one_hot
```

```python
para = ['I am Rao Umair Ali',
        'I am a lazy person',
        'Just doing nothing at all',
        'Always complaning about meal']
```

```python
para
```

```python
voc_size = 10000
```

```python
one_hot_emd = [one_hot(kuchbhi, voc_size) for kuchbhi in para]
one_hot_emd
```

```python
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 10
embedd = pad_sequences(one_hot_emd, maxlen=sent_length, padding='pre')
embedd
```

```python
dim = 12
```

```python
model =Sequential()
model.add(Embedding(voc_size, 12, input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedd)
```
